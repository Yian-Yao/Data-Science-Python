## Question 1

```python
def read_add(file_name, ticker):
    # Pass the absolute path into read function and add label column
    df = pd.read_csv(get_path(file_name, ticker))
    df['True Label'] = df['Return'].apply(lambda x: "+" if x >= 0 else "-")
    return df

def get_path(file_name, ticker):
    # Get absolute file path
    here = os.path.abspath(file_name)
    input_dir = os.path.abspath(os.path.join(here, os.pardir))
    ticker_file = os.path.join(input_dir, ticker + '.csv')
    return ticker_file
```

```python
# 1)
# Initialize ticker and file names
ticker_y = 'Y'
ticker_s = 'SPY'
file_name = 'Assignment 2'

df_y = read_add(file_name, ticker_y)
df_s = read_add(file_name, ticker_s)

# Show the table
df_y
```

```python
# 2)
def calc_prop(df, ticker):
    # Select first 3 years
    df_3 = df[df['Date'] <= '2018-12-31']
    
    # calculate the number of up day and down day
    up_day = len(df_3[df_3['True Label'] == '+'])
    down_day = len(df_3[df_3['True Label'] == '-'])
    
    # Calculate probability of up day
    prob = up_day / len(df_3)
    print(f'The probability p that the next day is a ”up” day of {ticker} is {round(prob* 100, 2)}%')
    # Retrun first 3 years data frame for following questions
    return df_3

df_3y = calc_prop(df_y, ticker_y)
df_3s = calc_prop(df_s, ticker_s)
```

```python
# 3)
def k_consecutive_down_days(df, ticker):
    # Pair k and the pattern
    for k, pattern in zip([1, 2, 3], [['-+', '--'], ['--+', '---'], ['---+', '----']]):
        
        # Covert each element in column into str and join them
        str_label = ''.join(str(x) for x in df['True Label'])
        
        # Use str.count(pattern) count required pattern
        prob = str_label.count(pattern[0]) / (str_label.count(pattern[0]) + str_label.count(pattern[1]))
        print(f'The probability of that after seeing {k} consecutive ”down days” the next day is an ”up '
              f'day" of {ticker} is {round(prob* 100, 2)}%')      
```

```python
k_consecutive_down_days(df_3y, ticker_y)
k_consecutive_down_days(df_3s, ticker_s)
```

```python
# 4)
def k_consecutive_up_days(df, ticker):
    # Pair k and the pattern
    for k, pattern in zip([1, 2, 3], [['+-', '++'], ['++-', '+++'], ['+++-', '++++']]):
        # Covert each element in column into str and join them
        str_label = ''.join(str(x) for x in df['True Label'])
                
        # by using count('+-')/(count('+-') + count('++'))
        prob = str_label.count(pattern[0]) / (str_label.count(pattern[0]) + str_label.count(pattern[1]))
        print(f'The probability of that after seeing {k} consecutive ”up days” the next day is an ”down '
              f'day" of {ticker} is {round(prob* 100, 2)}%')
```

```python
k_consecutive_up_days(df_3y, ticker_y)
k_consecutive_up_days(df_3s, ticker_s)
```

```python
# 1)
def predict_label(W, df_3, df):
    # Get all history data for first prediction
    str_label = ''.join(str(x) for x in df_3['True Label'])

    # Start from the last 4 day in 2018
    train_s = df[df['Date'] >= '2018-12-26']

    # Initialise the prediction set
    prediction = ''

    for i in range(4, len(train_s)):
        curr_seq = ''.join(str(x) for x in train_s['True Label'][i-W:i])
        
        # Create pattern list for "seq, +" and "seq, -"
        seq_up = curr_seq + '+'
        seq_down = curr_seq + '-'

        # Find the count for 2 list
        seq_up_count = str_label.count(seq_up)
        seq_down_count = str_label.count(seq_down)

        # Since our default probability of positive is 52.92% > 50%
        if seq_up_count >= seq_down_count:
            prediction += '+'
        elif seq_up_count < seq_down_count:
            prediction += '-'
    return prediction
```

```python
y_2 = predict_label(2, df_3y, df_y)
y_2_list = [str(i) for i in y_2]

y_3 = predict_label(3, df_3y, df_y)
y_3_list = [str(i) for i in y_3]

y_4 = predict_label(4, df_3y, df_y)
y_4_list = [str(i) for i in y_4]
```

```python
s_2 = predict_label(2, df_3s, df_s)
s_2_list = [str(i) for i in s_2]

s_3 = predict_label(3, df_3s, df_s)
s_3_list = [str(i) for i in s_3]

s_4 = predict_label(4, df_3s, df_s)
s_4_list = [str(i) for i in s_4]
```

```python
# Assign predition to DataFrame
df_y45 = df_y[df_y['Date'] >= '2019-01-02'].copy()
df_y45.loc[:,'W2 Pred Label'] = y_2_list
df_y45.loc[:,'W3 Pred Label'] = y_3_list
df_y45.loc[:,'W4 Pred Label'] = y_4_list

df_s45 = df_s[df_s['Date'] >= '2019-01-02'].copy()
df_s45.loc[:,'W2 Pred Label'] = s_2_list
df_s45.loc[:,'W3 Pred Label'] = s_3_list
df_s45.loc[:,'W4 Pred Label'] = s_4_list
```

```python
# 2)
def predicted_validation(df, stock):
    for W in range(2,5):
        # total accuracy
        validation_list = []
        validation_list = [1 if df['True Label'][i]==df[f'W{W} Pred Label'][i] else 0 for i in range(df.index[0], df.index[-1]+1)]
        df[f'W{W} validation'] = validation_list
        percent = df[f'W{W} validation'].sum()/len(df)
        print(f'The total accuracy for stock {stock} W{W} is {round(percent* 100, 2)}%')
        
        # accuracy on "-"
        validation_list = []
        validation_list = [1 if (df['True Label'][i]=='-') & (df[f'W{W} Pred Label'][i]=='-') else 0 for i in range(df.index[0], df.index[-1]+1)]
        df[f'W{W} validation'] = validation_list
        percent = df[f'W{W} validation'].sum()/df['True Label'].values.tolist().count('-')
        print(f'The accuracy on "-" for stock {stock} W{W} is {round(percent* 100, 2)}%')
        
        # accuracy on "+"
        validation_list = []
        validation_list = [1 if (df['True Label'][i]=='+') & (df[f'W{W} Pred Label'][i]=='+') else 0 for i in range(df.index[0], df.index[-1]+1)]
        df[f'W{W} validation'] = validation_list
        percent = df[f'W{W} validation'].sum()/df['True Label'].values.tolist().count('+')
        print(f'The accuracy on "+" for stock {stock} W{W} is {round(percent* 100, 2)}%', '\n')
    
    return df
```

```python
df_s45 = predicted_validation(df_s45, 'SPY')
df_y45 = predicted_validation(df_y45, 'Y')
```

```python
# 1) compute ensemble labels for year 4 and 5 for both your stock and S&P-500.
def ensemble_predict(df):
    label_list = df[['W2 Pred Label', 'W3 Pred Label', 'W4 Pred Label']].values.tolist()
    ensemble_label = ['+' if x.count('+') >  x.count('-') else '-' for x in label_list]
    df['Ensemble Pred Label'] = ensemble_label
    return df
```

```python
# 2) 
def ensemble_validation(df, stock):
    # toatal ensemble accuracy
    validation_list = [1 if df['True Label'][i]==df['Ensemble Pred Label'][i] else 0 for i in range(df.index[0], df.index[-1]+1)]
    df['Ensemble validation'] = validation_list
    percent = df['Ensemble validation'].sum()/len(df)
    print(f'The accuracy of ensemble predicted for stock {stock} is {round(percent* 100, 2)}%')
    
    # ensemble accuracy for "-"
    validation_list = [1 if (df['True Label'][i]=='-') & (df['Ensemble Pred Label'][i]=='-') else 0 for i in range(df.index[0], df.index[-1]+1)]
    df['Ensemble validation'] = validation_list
    percent = df['Ensemble validation'].sum()/df['True Label'].values.tolist().count('-')
    print(f'The accuracy for “-” of ensemble predicted for stock {stock} is {round(percent* 100, 2)}%')
    
    # ensemble accuracy for "+"
    validation_list = []
    validation_list = [1 if (df['True Label'][i]=='+') & (df['Ensemble Pred Label'][i]=='+') else 0 for i in range(df.index[0], df.index[-1]+1)]
    df['Ensemble validation'] = validation_list
    percent = df['Ensemble validation'].sum()/df['True Label'].values.tolist().count('+')
    print(f'The accuracy for “+” of ensemble predicted for stock {stock} is {round(percent* 100, 2)}%', '\n')
    
    return df
```

```python
df_s45 = ensemble_validation(df_s45, 'SPY')
df_y45 = ensemble_validation(df_y45, 'Y')
```

```python
# calculate TP/FP/TN/FN
def compute_stats(df, label, true_label, pred_label, stats, method, stock):
    # toatal ensemble accuracy
    validation_list = [1 if (df['True Label'][i]==true_label) & (df[label][i]==pred_label) else 0 for i in range(df.index[0], df.index[-1]+1)]
    count = sum(validation_list)
    print(f'The {stats} for {method} of stock {stock} is {count}')
    return count
```

```python
for W in ['W2', 'W3', 'W4', 'Ensemble']:
    tp_s = compute_stats(df_s45, f'{W} Pred Label', '+', '+', 'TP', W, 'SPY')
    fp_s = compute_stats(df_s45, f'{W} Pred Label', '-', '+', 'FP', W, 'SPY')
    tn_s = compute_stats(df_s45, f'{W} Pred Label', '-', '-', 'TN', W, 'SPY')
    fn_s = compute_stats(df_s45, f'{W} Pred Label', '+', '-', 'FN', W, 'SPY')
    # calculate TPR/TNR
    tpr_s = tp_s / (tp_s + fn_s)
    tnr_s = tn_s / (tn_s + fp_s)
    print(f'The TPR for {W} of stock SPY is {tpr_s}')
    print(f'The TNR for {W} of stock SPY is {tnr_s}')
    print('\n')
```

```python
for W in ['W2', 'W3', 'W4', 'Ensemble']:
    tp_y = compute_stats(df_y45, f'{W} Pred Label', '+', '+', 'TP', W, 'Y')
    fp_y = compute_stats(df_y45, f'{W} Pred Label', '-', '+', 'FP', W, 'Y')
    tn_y = compute_stats(df_y45, f'{W} Pred Label', '-', '-', 'TN', W, 'Y')
    fn_y = compute_stats(df_y45, f'{W} Pred Label', '+', '-', 'FN', W, 'Y')
    tpr_y = tp_y / (tp_y + fn_y)
    tnr_y = tn_y / (tn_y + fp_y)
    print(f'The TPR for {W} of stock Y is {tpr_y}')
    print(f'The TNR for {W} of stock Y is {tnr_y}')
    print('\n')
```

```python
# Calculate trade for best W W4 and Ensemble
def calculate_money(df, label):
    current = 100
    money = []
    for i in range(len(df)):
        if df[label].values[i] == '+':
            current += current * df['Return'].values[i]
        money.append(current)
    return money

# Calculate trade for buy and hold
def buy_n_hold(df):    
    current=100
    bnh_money = []
    for i in range(len(df)):
        current += current * df['Return'].values[i]
        bnh_money.append(current)
    return bnh_money
```
